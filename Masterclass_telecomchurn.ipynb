```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
import numpy as np

corr= df.corr()
masker = np.triu(np.ones_like(corr, dtype=bool))

cmap = sns.diverging_palette(230, 20, as_cmap=True)

sns.heatmap(corr, annot= True, mask= masker, cmap=cmap )
```
